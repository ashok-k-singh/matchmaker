# Load to AWS the matchdata table used to train the model

## Load dependencies and import data from S3 bucket

```python
import pandas as pd
```

```python
df = pd.read_csv('https://aws-matchmaker-bucket.s3-us-west-2.amazonaws.com/SpeedDatingPreprocessed.csv')
```

```python
df.head()
```

## Connect to AWS RDS database and load the table

Predecessor tasks: create the database via AWS Management Console.

```python
from sqlalchemy import create_engine
import config
engine = create_engine(f"postgresql+psycopg2://{config.pguser}:{config.pgpassword}@{config.pghost}:{config.pgport}/{config.pgdatabase}")
```

Predecessor tasks: create table via PGAdmin

```python
df.to_sql('matchdata', engine, if_exists='append',index=False)
```
